## knn ##

** Adjusting Religion **

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
knn = KNeighborsClassifier(n_neighbors=4, 
                            weights='uniform')
X_ = Ohio[['Education_Category','Income_Category', 'Latitude', 'Longitude', 'Marital_Status', 'Party', 'Party_Home']]
Xs_ = ss.fit_transform(X_)
print(len(Xs_))

y_ = Ohio['Religion']

Xs_ = pd.DataFrame(Xs_, columns=['Education_Category','Income_Category', 'Latitude', 'Longitude', 'Marital_Status', 'Party', 'Party_Home'])
all_ = pd.DataFrame(Xs_)
all_['Religion'] = y_

all_.dropna(inplace=True)
all_dropped = pd.DataFrame(all_[(all_['Religion'] != 8.0) & (all_['Religion'] != 9.0)])

y_ = all_dropped['Religion']
Xs_ = all_dropped.drop('Religion', axis=1)

scores = cross_val_score(knn, Xs_, y_, cv=5)

print(scores)

knn.fit(Xs_, y_)

knn.predict(Ohio[['Education_Category','Income_Category', 'Latitude', 'Longitude', 'Marital_Status', 'Party', 'Party_Home']])

new_Xs = ss.transform(Ohio[['Education_Category','Income_Category', 'Latitude', 'Longitude', 'Marital_Status', 'Party', 'Party_Home']])

Religion_pred = knn.predict(new_Xs)

Ohio['Religion_pred'] = Religion_pred

len(Religion_pred)
```

```python
pd.get_dummies(Ohio['Religion_pred'])
Ohio = pd.concat([Ohio, pd.get_dummies(Ohio['Religion_pred'])], axis = 1)
Ohio.rename(columns={1.0:'Protestant_D_RP', 2.0:'Catholic_D_RP', 3.0:'Jewish_D_RP', 
                     4.0:'EasternGreekOrthodox_D_RP', 5.0:'Christian_D_RP', 
                     6.0:'Asian_D_RP', 7.0:'Islamic_D_RP'}, inplace=True)

Ohio['Religion_pred'].value_counts()
```
