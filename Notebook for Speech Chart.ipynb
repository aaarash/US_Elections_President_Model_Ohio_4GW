```python
import matplotlib.pyplot as plt
```

```python
import nltk
import pandas as pd
import string
import numpy as np
from matplotlib import style
style.use("ggplot")
```

```python
messages1 = [line.rstrip() for line in open("data/presidential-campaign-trail/Trump.txt", encoding='utf-8')]
messages2 = [line.rstrip() for line in open("data/presidential-campaign-trail/Clinton.txt", encoding='utf-8')]
messages3 = [line.rstrip() for line in open("data/presidential-campaign-trail/Romney.txt", encoding='utf-8')]
messages4 = [line.rstrip() for line in open("data/presidential-campaign-trail/Obama.txt", encoding='utf-8')]
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
# install Vader and make sure you download the lexicon as well
sid = SentimentIntensityAnalyzer()
labels = ["Positive","Neutral", "Negative"]
colors = ["forestgreen", "beige","firebrick"]
```

```python
summary = {"positive":0,"neutral":0,"negative":0}

for x in messages1: 
    ss = sid.polarity_scores(x)
    if ss["compound"] == 0.0: 
        summary["neutral"] +=1
    elif ss["compound"] > 0.0:
        summary["positive"] +=1
    else:
        summary["negative"] +=1
print(summary)            
```

```python
trump = pd.DataFrame({
    "Sentiment":["Positive","Neutral","Negative"],
    "Trump":[34,56,5],
})

trump = trump.set_index("Sentiment")
plt.title("Trump Speech")
plt.pie(trump, labels = labels, explode = (0,0.069,0), colors = colors, autopct = "%#1.1f%%", shadow =True,startangle = 140)
plt.axis("equal")
plt.savefig("Charts/Trump Speech.png")
plt.tight_layout()
plt.show()
```
